# **SOCIAL MEDIA DATA ANALYSIS**
This data is related to analytical data that shows the performance and measures the performance of content or posts created by Instagram users. This performance is referred to as insight which is based on the type of content created. From the insight obtained, whether it can become a business for the owner of the profile or not.

# **FEATURES DESCRIPTION**

- Month : month when uploading content
- Date : detailed information regarding upload times 
- Profile Name : a unique name given to each user on the platform ('Orange_id', 'Strawberry Spot')
- Channel : name of the social media application used
- Content Type :the type of content or posts that users can upload ('carousel', 'video', 'image', nan)
- Likes : an action taken by a user to show that they like or appreciate a post.
- Comments : the user's action to leave a text message in response or in response to a post.
- Labels : ('ORGANIC', 'PROMOTED')
        - ORGANIC : that appear naturally in users' feeds or pages without any payment to increase the reach of their content.
        - PROMOTED : the content has been promoted or advertised by another user or entity.
- Interactions :interaction includes likes, comments, direct messages, shares, saves, poll responses, visiting your profile, clicking links, mentions, tagging, and more (numeric)
- Insight - Reach : how many unique users are served your content every day (numeric)
- Insight - Impression :measure the number of times users view your content (numeric)
- Insight - Saves : allows users to save posts they find interesting
- Insight - Engagement : engagement analytics that empower you to improve your Instagram for Business strategy (numeric)
- Insight - Video Views :the number of people who have watched a video on the platform
